```python
# 실습 1: 생일 기온 시각화

import csv
import matplotlib.pyplot as plt

f = open('seoul_temp.csv','r',encoding='cp949')
data = csv.reader(f)

next(data)
result1 = []
result2 = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '10' and row[0].split('-')[2] == '20':
            result1.append(float(row[-1]))
            result2.append(float(row[-2]))

plt.plot(result1)
plt.plot(result2)
plt.show()
```

```python
# 실습 2: 최고 기온 시각화

import csv
import matplotlib.pyplot as plt

f = open('seoul_temp.csv','r',encoding='cp949')
data = csv.reader(f)

next(data)
jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        if month == '08':
            aug.append(float(row[-1]))

plt.hist(jan, bins=100, color='b')
plt.hist(aug, bins=100, color='r')
plt.show()
```
